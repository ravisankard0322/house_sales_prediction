```python
import pandas as pd
import numpy as np
#https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv")
df
```

```python
df.info()
```

```python
numeric_data = df.select_dtypes(include=['int64','float64'])
numeric_data 
```

```python
x = numeric_data.iloc[:,0:3].values
y = numeric_data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split   #(75%:25% split)
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0) 
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred 
```
